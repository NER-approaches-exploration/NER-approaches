```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame.from_dict({
    'method': ['Natasha', 'Dictionary', 'CRF', 'AI-Forever RuBERT', 'Google BERT', 'Google BERT multilingual'],
    'CoNLL 2003': [0, 0.3758, 0.6655, 0.7871, 0.8980, 0.8990],
    'BC5CDR': [0, 0.2450, 0.4498, 0.7275, 0.8005, 0.8148],
    'Collection3': [0.5692, 0.2740, 0.5405, 0.9433, 0.8496, 0.9696],
    'FactRuEval 2016': [0.1879, 0.6767, 0.4142, 0.9464, 0.8226, 0.9604]
})
df.set_index('method', inplace=True)
print(df)
```

```python
df.to_excel('results_f1.xlsx')
```

```python
df_long = df.reset_index().melt(id_vars='method', var_name='dataset', value_name='f1')

plt.figure(figsize=(16, 10))
sns.barplot(data=df_long, x='dataset', y='f1', hue='method')
plt.title('F1 Score по моделям и датасетам', fontsize=20)
plt.xlabel('Датасет', fontsize=16)
plt.ylabel('F1 Score', fontsize=16)
plt.ylim(0, 1)

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.legend(title='Модель', title_fontsize=15, prop={'size': 13})
plt.subplots_adjust(right=1.75)

plt.savefig('f1_scores.png', bbox_inches='tight', dpi=450)

plt.show()
```
